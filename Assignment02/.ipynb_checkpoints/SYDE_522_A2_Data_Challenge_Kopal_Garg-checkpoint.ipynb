```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.model_selection import KFold,cross_val_score,train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn import metrics
import umap
```

Load the train and test set

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

Function for PCA and training a SVM on PCA components. Used a randomized grid search to find optimal hyperparameters. 

```python
def pca_svc_rbf(train, test):
    
    X = train.drop("label", axis=1)
    y = train["label"].values
    # normalization 
    test = test/255.0 
    X_train = X/255.0
    # split X s.t. we have 90% for training and 10% for validation
    X_train, X_valid, y_train, y_valid = train_test_split(X_train, y, train_size=0.90)
    # dimensionality reduction using PCA. 33 PCs were chosen after much trial and error
    pca = PCA(n_components=33, whiten=True)
    X_train = pca.fit_transform(X_train)
    X_valid = pca.transform(X_valid)
    test = pca.transform(test)  
    # SVM on Gaussian kernel. Hyperparameters were chosen using a grid-search (runtime >1hr, so not included in function)
    svc = SVC(kernel="rbf", C = 10, gamma= 0.05, random_state=50)
    svc.fit(X_train, y_train)
    pred = svc.predict(test)

    return pred
```

Kaggle submission - best score 0.98375

```python
test = pd.read_csv('test.csv')
test_submission = pd.DataFrame({'id': [i for i in range(test.shape[0])],
                                'label': pca_svc_rbf(train, test)})
test_submission.to_csv('KopalGarg-SYDE522-A2-Submission.csv', index=False)
```

Grid search - Not included in function as it takes >1 hours to run

```python
svc_params = [{'kernel': ['rbf'],
             'C': [10,100,1000],
             'gamma': [0.1, 0.01,0.05, 0.001] }]
grid = RandomizedSearchCV(SVC(kernel='rbf', cache_size=6000, random_state=50), svc_params, n_iter=30, 
                           scoring='accuracy')
grid.fit(X_train, y_train)

bestRM=grid.best_estimator_

y_hat_valid = bestRM.predict(X_valid)
print("Accuracy: ", accuracy_score(y_valid, y_hat_valid))
```
