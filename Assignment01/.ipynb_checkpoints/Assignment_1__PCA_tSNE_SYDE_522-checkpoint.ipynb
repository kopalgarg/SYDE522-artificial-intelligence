# [NO MARKS] PCA Warming Up (MUST READ)

Some code to illustrate examples of PCA using sklearn library.

**Let's create some random `5d` data**

```python
import numpy as np
from sklearn.decomposition import PCA

# 100 points of 5d data
data = np.random.rand(100, 5)
```

**Lets convert this `5d` data to `2d` using `PCA`**


```python
# n_components=2 because I want to convert 5d data to 2d i.e, d=5 and k=2 based on the notation in Tutorial.
pca = PCA(n_components=2)
pca.fit(data)
```

In code above when we call `fit`, it populates two things in `pca`:
1. `mean_`
2. `components_`

```python
# mean of the input data (per dimension) used to zeroying the mean
pca.mean_
```

```python
# basis vectors: k by d projection marix (this is the transposed of projection matrix based on notation in the Tutorial.)
pca.components_
```

Now we are ready to transform `5d` from `data` into `2d` using following code

```python
data_to_reduce = data[:10]
reduced_data = np.dot(data_to_reduce - pca.mean_, pca.components_.T)

# reduced data from 5d to 2d
reduced_data.shape
```

You can accomplish the same using `transform` function provided in `pca`

```python
pca.transform(data_to_reduce).shape
```

Time for inverse transform or changing `2d` data back to `5d`

Compression --> Decompression

```python
decompressed_data = np.dot(reduced_data, pca.components_)+pca.mean_
decompressed_data
```

```python
# same can we accomplished using inverse_transform
pca.inverse_transform(pca.transform(data_to_reduce))
```

```python
# Lets find compression decompression error (absolute mean error)
np.sum(np.abs(data_to_reduce - decompressed_data))/data_to_reduce.size
```

## 1-a) [10 marks]
Show that minimizing the reconstruction loss in PCA i.e,

\begin{equation}
\begin{aligned}
& \underset{\mathbf{W}}{\text{arg min}} &&   \| \mathbf{X}-\mathbf{X}\mathbf{W}\mathbf{W}^T \| ^2_F,  \quad \mathbf{W}^T \mathbf{W}=\mathbf{I}_k, \quad (1)
\end{aligned}
\end{equation}

is equivalent to maximizing variance in all orthogonal projction directions i.e,

$$
\underset{\mathbf{W}}{\text{arg max}} \quad \textrm{Tr}\{\mathbf{W}^{T}\mathbf{X}^{T}\mathbf{X}\mathbf{W}\}\quad
\textrm{s.t.} \quad  \mathbf{W}^T \mathbf{W}=\mathbf{I}_k, \quad (2)
$$
where $ \textrm{Tr}$ is trace function, $\| .\|_F$ is frobenius norm, $\mathbf{X}$ is $n$ by $d$ data matrix ($n$ number of data points and $d$ dimensionality), $\mathbf{W}$ is a $d$ by $k$  projection matrix, and $\mathbf{I}_k$ is $k$ by $k$ identity matrix.

Hint 1: Try to expand the frobenus norm in relation (1) and note that $\mathbf{W}^T \mathbf{W}=\mathbf{I}_k$.

Hint 2: In order to show this you may use Trace of a product and Cyclic property of the Trace function. You can read about Cyclic property of Trace function [here](https://en.wikipedia.org/wiki/Trace_(linear_algebra)). <br>
<br>


Answer to 1-a):

<font color='blue'>
    
$X$ is an $n$ by $d$ data matrix. $\Sigma$ is a covariance matrix given by:

\begin{equation}
\begin{aligned}
& \Sigma = \frac{X^T X}{n-1}
\end{aligned}
\end{equation} 

Minimizing the reconstruction error between $X$ and its reconstruction $ XWW^T$:

\begin{equation}
\begin{aligned}
& \underset{W}{\text{arg min}} &&   \| X-XWW^T \| ^2_F,  \quad W^T W=I_k
\end{aligned}
\end{equation}    
    
\begin{equation}
\begin{aligned}
&=tr([X-XWW^T]^T[X-XWW^T])
\end{aligned}
\end{equation}  
    
\begin{equation}
\begin{aligned}
&=tr([X^T-X^{T}W^{T}W][X-XWW^T])
\end{aligned}
\end{equation}  

\begin{equation}
\begin{aligned}
&=tr(X^TX-2X^TWW^TX+X^TWW^TWW^TX)
\end{aligned}
\end{equation}  

Since $W^TW=I_k$
    
\begin{equation}
\begin{aligned}
&=tr(X^TX-2X^TWW^TX+X^TWW^TX)
\end{aligned}
\end{equation}  

\begin{equation}
\begin{aligned}
&=tr(X^TX-X^TWW^TX)
\end{aligned}
\end{equation}  

Since \Sigma is a covariance matrix:

\begin{equation}
\begin{aligned}
& \Sigma = \frac{X^T X}{n-1}
\end{aligned}
\end{equation} 

\begin{equation}
\begin{aligned}
&=constant_1-constant_2 . W \Sigma W^T
\end{aligned}
\end{equation}  

    
Maximizing variance in all orthogonal projction directions is given by:

$$
\underset{W}{\text{arg max}} \quad \textrm{Tr}\{W^{T}X^{T}XW\}\quad
\textrm{s.t.} \quad  W^T W=I_k, \quad (2)
$$
    
\begin{equation}
\begin{aligned}
&=tr(WX^TXW^T)
\end{aligned}
\end{equation}  

\begin{equation}
\begin{aligned}
&=constant_3 . W\Sigma W^T
\end{aligned}
\end{equation}
    
Therefore, the two optimization problems are equivalent.
    
</font>

Show that projection matrix in PCA i.e, $\mathbf{W}$ is rotation invariant .
Hint: You need to show that if we replace $\mathbf{W}$ with $\mathbf{W}\mathbf{R}$ where $\mathbf{R}$ is a $k$ by $k$ rotation matrix (rotation matrix is a square orthohonal matrix such that $\mathbf{R}^T\mathbf{R}=\mathbf{R}\mathbf{R}^T=\mathbf{I}_k$) then the objective value in relation (2) is the same.

Hint: In order to show this you may use Trace of a product and Cyclic property of the Trace function. You can read about Cyclic property of Trace function [here](https://en.wikipedia.org/wiki/Trace_(linear_algebra)).

Answer to 1-b):

<font color='blue'>

R is a $k$ by $k$ rotation matrix, such that $R$$R^T$ = $R^T$$R$ = $I_k$

Maximizing variance in all orthogonal projction directions is given by:

$$
\underset{W}{\text{arg max}} \quad \textrm{Tr}\{W^{T}X^{T}XW\}\quad
\textrm{s.t.} \quad  W^T W=I_k, \quad (2)
$$ 

\begin{equation}
\begin{aligned}
&=tr(WX^TXW^T)
\end{aligned}
\end{equation}  

\begin{equation}
\begin{aligned}
&=constant . W\Sigma W^T
\end{aligned}
\end{equation}

If $W$ is replaced with $WR$ where $R$ is a $k$ by $k$ rotation matrix:
    
\begin{equation}
\begin{aligned}
&=tr(WRX^TXW^TR^T)
\end{aligned}
\end{equation}  

\begin{equation}
\begin{aligned}
&=constant . WR\Sigma W^TR^T
\end{aligned}
\end{equation}    

Because $R$$R^T$ = $R^T$$R$ = $I_k$:
    
\begin{equation}
\begin{aligned}
&=constant . W\Sigma W^T
\end{aligned}
\end{equation}  
 
Therefore, the projection matrix, $W$ in PCA is rotation invariant.

## 1-c) [10 marks] 
Assume we have the data matrix 
$\mathbf{X}$ $n$ by $d$  ($n$ number of data points and $d$ dimensionality), and the projection martix $\mathbf{W}$ ($d$ by $k$) from PCA algorithm. Explain in detail how we can use this $\mathbf{W}$ to denoise $\mathbf{X}$? <br>

write the mathematical exparession for calcuating the denoised data and explain role of $k$ in denoising.

Answer to 1-c):

<font color='blue'>

PCA aims to reduce the dimensionality of a data matrix with correlated or irrelavant variables, while maintaining maximum variation present in the data. This is done by transforming $X$ into $W$, the projection (or eigenvector) matrix, such that the principle components (PCs) are uncorrelated and the top $k$ eigenvalues explain most of the variance present in the data.

An orthogonal linear transformation is used to find a projection of all data into $k$ dimensions. This involves computing the mean for every dimension of the data matrix, followed by finding the covariance matrix, computing eigenvectors ($v$) and corresponding eigenvalues ($\lambda$). The eigenvectors are then sorted by decreasing eigenvalues and top k eigenvectors are retained to form a $d$ by $k$ matrix, W. Eigenvalues denote the amount of variability captured along the dimension, s.t. $\lambda_1 \geq \lambda_2 \geq \lambda_3 \geq ...\lambda_n$ . 
    
Because noise affects all eigenvalues, choosing just the top $k$ eigenvectors reduces the influence of noise on the data while also preserving necessary information. So, if we reconstruct the data using just the largest subset of principal components, we are keeping the signal and removing the noise. $XW$ results in a $n$ by $k$ matrix as PCs of least signigicance are removed. Multipying $XW$ by the transpose of the projection matrix, $W^T$, gives us the data in its inital form. The expression for calculating the de-noised data is:
    
\begin{equation}
\begin{aligned}
& X = XWW^T
\end{aligned}
\end{equation}  

    
</font>

Generate a random matrix sampled from normal distribution with mean equal to 1 and standard deviation equal to 20 with the size of X. Name this matrix noise and add this to the X to get noisy_X. Then run the cell to visualize some of the noisy images.

```python
# write your code here
import numpy as np

np.random.seed(2406)
mu, sigma = 1, 20
noise= np.random.normal(mu, sigma, size = X.shape)
noisy_X = X + noise
fig, ax = plt.subplots(1, 7, figsize=(9, 5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(7):
    ax[i].imshow(noisy_X[i].reshape(62, 47), cmap='bone')
```

##3-b) [5 marks]
Denose the images using PCA algorithm. Choose $k$ such that 80 % of variance in data is captured. name the denoised data "filtered".

```python
# write the denoising code here name the denoised data "filtered"
pca = PCA(0.80).fit(noisy_X)
components = pca.transform(noisy_X)
filtered = pca.inverse_transform(components)
```

Run the next cell to visualize some of the denoised images.

```python
fig, ax = plt.subplots(1, 7, figsize=(9, 5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(7):
    ax[i].imshow(filtered[i].reshape(62, 47), cmap='bone')
```

## Problem 4: tSNE [15 marks]
In the following the digit dataset is loaded. The images and corresponsing labels are loaded in data and label variables

```python
from sklearn.datasets import load_digits
digits = load_digits()
data = digits.data
label = digits.target
data.shape
label.shape
```

Use PCA and tSNE to obtain two-dimensional repesentations for this dataset. Use plt.scatter function to plot the 2-D representation of this dataset obtained by both PCA and tSNE method. Use labe information to assign distinct colors for data points in different classes. Based on the generated graphs, which one is a better method for visualization?

```python
# Write your code to apply PCA here
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import time
time_start = time.time()
pca = PCA(2)
projected = pca.fit_transform(digits.data)
projected.shape
print(time.time()-time_start)
```

```python
# Write your code to plot the scatter plot for 2-D represntation obtained from PCA here
plt.scatter(projected[:, 0], projected[:, 1],
            c = label, edgecolor='none', alpha=0.5)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
# Write your code to apply TSNE here
from sklearn.manifold import TSNE
import time
time_start = time.time()
X_embedded = TSNE(n_components=2, random_state = 0).fit_transform(digits.data)
X_embedded.shape
print(time.time()-time_start)
```

```python
# Write your code to plot the scatter plot for 2-D represntation obtained from TSNE here
plt.scatter(X_embedded[:, 0], X_embedded[:, 1],
            c=label, alpha=0.5)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

Wrtie your conclusion about employing PCA and tSNE for data visualization

<font color='blue'>
    
PCA is an unsupervised linear dimensionality reduction and data visualization technique for high dimensional data. It preserves the global structure of the data and is sensitive to outliers. By comparing the visualisations produced by the two models, we can see that PCA does not do a good job in differentiating the digits. This is because PCA is a linear technique , and is unable to capture non-linear dependencies. T-SNE is an unsupervised non-linear technique that can capture the structure of trickier non-linear manifolds. It preserves the local structure of the data and can handle outliers. The scatter plot for TSNE shows several well-separated clusters. It does a better job at categorizing the digits. However, t-SNE (6.657s) takes considerably longer to execute on the same sample size of data than PCA (0.006s).
    
    
</font>
